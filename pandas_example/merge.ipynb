```python
import pandas as pd

sample_1 = pd.read_excel('./files/sample_1.xlsx', header=1, skipfooter=2, usecols='A:C')
sample_1.head(3)
```

# 데이터 통합 - 옆으로

```python
code_master = pd.read_excel('./files/sample_codemaster.xlsx')
code_master
```

```python
# left merge
# 왼쪽 테이블을 기준으로 매칭되는 데이터가 없을 경우 NaN
sample_1_code = pd.merge(left=sample_1, right=code_master, how='left', left_on='국적코드', right_on='국적코드')

sample_1_code
```

```python
# inner merge
# 양쪽 테이블의 기준 칼럼이 서로 일치하는 경우에만 데이터 통합
sample_1_code_inner = pd.merge(left=sample_1, right=code_master, how='inner', left_on='국적코드', right_on='국적코드')

sample_1_code_inner
```

# 데이터 통합 - 아래로

```python
sample_1_code['기준년월'] = '2019-11'

sample_2 = pd.read_excel('./files/sample_2.xlsx', header=1, skipfooter=2, usecols='A:C')
sample_2_code = pd.merge(left=sample_2, right=code_master, how='left', left_on='국적코드', right_on='국적코드')
sample_2_code['기준년월'] = '2019-12'

sample_2_code
```

```python
sample_1_code
```

```python
sample = sample_1_code.append(sample_2_code, ignore_index=True)
sample
```

```python
# to_excel은 openpyxl에 의존성이 있음
sample.to_excel('./files/sample.xlsx',index=False)
```

```python
sample.to_csv('./files/sample.csv', index=False)
```
