```python
import pandas as pd

sample = pd.read_excel('./files/sample.xlsx')
sample
```

```python
sample_pivot = sample.pivot_table(values='입국객수', index='국적명', columns='기준년월', aggfunc='mean')
# aggfunc 옵션:
#     mean / sum / min / median / max / count / nunique(중복을 제거한 후 개수)
sample_pivot
```

```python
# column값을 지정하지 않은 pivot table
sample_pivot_2 = sample.pivot_table(values='입국객수', index='국적명', aggfunc='max')
sample_pivot_2
```
