```python
# 그래프에서 한글을 표기하기 위한 글꼴 변경(윈도우, macOS에 대해 각각처리)
# jupyter에서 기본 글꼴이 한글을 지원하지 않아서 필요
from matplotlib import font_manager, rc
import platform

if platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname = path).get_name()
    rc('font', family = font_name)
elif platform.system() == 'Darwin':
    rc('font', family = 'AppleGothic')
else:
    print('Check your OS')
```

```python
import matplotlib.pyplot as plt

condition = (kto_total['국적'] == '중국')
kto_total_filter = kto_total[condition]
kto_total_filter.head(12)
```

```python
plt.plot(kto_total_filter['기준년월'], kto_total_filter['관광'])
plt.show()
```

```python
# 위에 그래프는 X, Y축을 알아보기 어렵다.

## 그래프 크기 조절
plt.figure(figsize=(12, 4))

## 그래프 내용 설정
plt.plot(kto_total_filter['기준년월'], kto_total_filter['관광'])

## 그래프 타이을, X, Y축 이름 달기
plt.title('중국 국적의 관광객 추이')
plt.xlabel('기준년월')
plt.ylabel('관광객수')

## X축 눈금값 설정
plt.xticks(['2010-01', '2011-01', '2012-01', '2013-01', '2014-01', '2015-01', '2016-01', '2017-01', '2018-01', '2019-01', '2020-01'])

plt.show()
```
