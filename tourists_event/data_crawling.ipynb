```python
# data from https://kto.visitkorea.or.kr/kor/notice/data/statis/profit/notice/inout/popup.kto
# 2010.01 ~ 2020.08

from selenium import webdriver
from selenium.webdriver.support.ui import Select

import os
import shutil
import time


driver = webdriver.Chrome('C:/Users/gram/Downloads/chromedriver_win32/chromedriver.exe')

start_year = 2010
end_year = 2020
months = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']

for yyyy in range(start_year, end_year + 1):
    url = 'https://kto.visitkorea.or.kr/kor/notice/data/statis/profit/notice/inout/popup.kto'
    driver.get(url)
    time.sleep(3)

    gubun_1 = driver.find_element_by_id('gubun_1')
    gubun_1_select = Select(gubun_1)
    gubun_1_select.select_by_value('2')

    gubun_2 = driver.find_element_by_id('gubun_2')
    gubun_2_select = Select(gubun_2)
    gubun_2_select.select_by_value('2')

    year = driver.find_element_by_id('yyyy')
    year_select = Select(year)
    year_select.select_by_value(str(yyyy))
    
    for mm in months:
        try:
            month = driver.find_element_by_id('mm')
            month_select = Select(month)
            month_select.select_by_value(mm)
            
            search = driver.find_element_by_xpath('/html/body/div[2]/div/form[1]/fieldset/div/div[2]/a')
            search.click()
            time.sleep(3)
            
            download = driver.find_element_by_xpath('/html/body/div[2]/div/div[2]/a[2]')
            download.click()
            time.sleep(5)
            
            shutil.copy("C:/Users/gram/Downloads/file.xls", f"C:/Users/gram/Desktop/Projects/data-with-code/thaad/files/kto_{yyyy}_{mm}.xls")
            os.remove("C:/Users/gram/Downloads/file.xls")
            
            print(f"{yyyy}_{mm}")
            time.sleep(3)
        except:
            print(f"{yyyy}_{mm} : fail")
            pass
```

```python
# data from http://kto.visitkorea.or.kr/kor/ktom/stat/stat.kto
# 2010.01 ~ 2020.08

from selenium import webdriver
from selenium.webdriver.common.keys import Keys

import os
import re
import shutil
import time


driver = webdriver.Chrome('C:/Users/gram/Downloads/chromedriver_win32/chromedriver.exe')

url = 'http://kto.visitkorea.or.kr/kor/ktom/stat/stat.kto'
driver.get(url)
time.sleep(3)

sub_menu = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div/aside/ul/li[1]/ul/li[9]/a')
sub_menu.click()
time.sleep(3)

option_time_month = driver.find_element_by_id('option-month')
option_time_month.click()
time.sleep(3)

start_year = driver.find_element_by_xpath('/html/body/div[2]/div[3]/div[1]/div[2]/div/div[2]/div[2]')
start_year.click()
start_year_selects = start_year.find_elements_by_tag_name('li')
start_year_selects[10].click()

start_month = driver.find_element_by_xpath('/html/body/div[2]/div[3]/div[1]/div[2]/div/div[2]/div[3]')
start_month.click()
start_month_selects = start_month.find_elements_by_tag_name('li')
start_month_selects[0].click()

# end_year = driver.find_element_by_xpath('/html/body/div[2]/div[3]/div[1]/div[2]/div/div[3]/div[2]')
# end_year.click()
# end_year_selects = end_year.find_elements_by_tag_name('li')
# end_year_selects[0].click()

end_month = driver.find_element_by_xpath('/html/body/div[2]/div[3]/div[1]/div[2]/div/div[3]/div[3]')
end_month.click()
end_month_selects = end_month.find_elements_by_tag_name('li')
end_month_selects[7].click()

country_option = driver.find_element_by_xpath('/html/body/div[2]/div[3]/div[3]/div[2]/div/div[2]/div[2]')
country_option.click()
countries = country_option.find_elements_by_class_name('group-option')
print(len(countries), 'countries')

country_html_list = []
for country in countries:
    html = country.get_attribute('innerHTML')
    country_html_list.append(html)

country_name_list = []
for idx, country in enumerate(country_html_list):
    try:
        country_name = re.sub('<.+?>', "", country, 0).strip()
        print(idx, country_name)
        country_name_list.append(country_name)
        
        country_input = driver.find_element_by_xpath('/html/body/div[2]/div[3]/div[3]/div[2]/div/div[2]/div[2]/div/ul/li/input')
        country_input.send_keys(Keys.BACKSPACE)
        time.sleep(2)
        country_input.send_keys(country_name)
        country_input.send_keys(Keys.RETURN)

        search_btn = driver.find_element_by_xpath('/html/body/div[2]/div[3]/div[4]/button[1]')
        search_btn.click()
        time.sleep(3)

        download_btn = driver.find_element_by_xpath('/html/body/div[2]/div[3]/div[5]/div[1]/button')
        download_btn.click()
        time.sleep(5)

        shutil.copy("C:/Users/gram/Downloads/외래객입국_목적별_국적별.xlsx", f"C:/Users/gram/Desktop/Projects/data-with-code/tourists_event/files/kto_{country_name}_201001_202008.xlsx")
        os.remove("C:/Users/gram/Downloads/외래객입국_목적별_국적별.xlsx")
        time.sleep(3)
    except:
        country_name = re.sub('<.+?>', "", country, 0).strip()
        print(country_name, ": fail")
        
print(len(country_name_list), "success")
print(country_name_list)
```
