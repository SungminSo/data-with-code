텍스트 내 출현하는 단어들의 출현 빈도수에 비례하는 코기로 표풀한 그래프

```python
import pandas as pd

raw_total = pd.read_excel('./files/crawling_jeju.xlsx')
raw_total['tags'][:3]
```

```python
tags_total = []

for tags in raw_total['tags']:
    tags_list = tags[2:-2].split("', '")
    for tag in tags_list:
        tags_total.append(tag.strip())
```

```python
print(len(tags_total))
```

```python
# Counter() 함수는 그룹 내 원소의 종류별 빈도수를 집계
from collections import Counter

tag_counts = Counter(tags_total)
```

```python
tag_counts.most_common(50)
```

```python
from matplotlib import font_manager, rc

import matplotlib.pyplot as plt
import seaborn as sns
import sys

# 한글 깨짐을 방지하기 위해 한글 글꼴 지정
if sys.platform in ['win32', 'win64']:
    font_name = 'malgun gothic'
elif sys.platform == 'darwin':
    font_name = 'AppleGothic'
    
rc('font', family=font_name)
```

```python
# 막대그래프
tag_counts_df = pd.DataFrame(tag_counts_selected.most_common(30))
tag_counts_df.columns = ['tags', 'counts']

plt.figure(figsize = (10, 8))
sns.barplot(x = 'counts', y = 'tags', data = tag_counts_df)
```

```python
from wordcloud import WordCloud

import matplotlib.pyplot as plt
import platform

# 한글 깨짐 방지를 위해서 wordcloud에서는 글꼴 경로를 지정해줘야 함
if platform.system() == 'Windows':  # 윈도우
    font_path = 'c:/Windows/Fonts/malgun.ttf'
elif platform.system() == 'Darwin':  # macOS
    font_path = '/Users/$USER/Library/Fonts/AppleGothic.ttf'
```

```python
wordcloud = WordCloud(font_path = font_path,
                     background_color = 'white',
                     max_words = 100,
                     relative_scaling = 0.3,
                     width = 800,
                     height = 400
                     ).generate_from_frequencies(tag_counts_selected)
plt.figure(figsize = (15, 10))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('./files/tag_wordcloud.png')
```
